```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
df = pd.read_csv('ulazni_podaci_vsl2.csv')
df.head()
```

```python
y = df['Klasifikacija']
X = df.drop(columns=['Naziv_proteina', 'Klasifikacija'], axis=1)
```

```python
feature_names = X.columns
df[feature_names]
```

```python
y.unique()
```

```python
y.replace("uređen", 1, inplace=True)
y.replace("neuređen", 0, inplace = True)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
X_test_index = X_test.index
```

```python
#STANDARDIZACIJA
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train) #samo ulazne podatke standardizujemo, ne standardizujemo ciljne promenljive(klase) i zato je ovde samo X_train
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)


X_train = pd.DataFrame(X_train, columns=feature_names).round(4)
X_test = pd.DataFrame(X_test, index=X_test_index, columns=feature_names).round(4)
```

```python
#Cuvanje trening i test podataka u CSV datoteke
trening_podaci = pd.concat([X_train, y_train], axis=1)
test_podaci = pd.concat([X_test, y_test], axis=1)
trening_podaci.to_csv('Trening podaci/VSL2/trening_podaci_KNN_vsl2.csv', index=False)
test_podaci.to_csv('Test podaci/VSL2/test_podaci_KNN_vsl2.csv', index=False)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import GridSearchCV

params_grid = { 'n_neighbors' : range(10,50,5),
                'weights' : ['uniform', 'distance'],
                'p' : [1,2]}

model = GridSearchCV(KNeighborsClassifier(), params_grid, cv=5, verbose=2)
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
best_model = model.best_estimator_
```

```python
best_model
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
y_pred_train = model.predict(X_train) 
confusion_matrix(y_train, y_pred_train)
```

```python
accuracy_score(y_train, y_pred_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# CSV fajl sa predikcijama za test skup
X_test = pd.DataFrame(X_test, index=X_test_index, columns=feature_names)

test_podaci = X_test.copy()  
test_podaci['Naziv_proteina'] = df.loc[X_test.index, 'Naziv_proteina']  
test_podaci['Stvarna vrednost'] = y_test  # Stvarne vrednosti
test_podaci['Predviđena vrednost'] = y_pred  # Predviđene vrednosti

test_podaci['Stvarna vrednost'] = test_podaci['Stvarna vrednost'].replace({0: 'neuređeno', 1: 'uređeno'})
test_podaci['Predviđena vrednost'] = test_podaci['Predviđena vrednost'].replace({0: 'neuređeno', 1: 'uređeno'})

test_podaci.insert(0, 'Naziv_proteina', test_podaci.pop('Naziv_proteina'))

# CSV fajl sa predikcijama za test skup
test_podaci.to_csv('Predikcije test skupa/VSL2/predikcije_test_skupa_KNN_vsl2.csv', index=False)
```
