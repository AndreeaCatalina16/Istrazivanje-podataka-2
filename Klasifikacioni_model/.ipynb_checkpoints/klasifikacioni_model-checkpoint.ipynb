```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
df = pd.read_csv('molske_frakcije_klasifikacija_spike_proteina_iupred3.csv')
df.head()
```

```python
df.any()
```

```python
y = df['Klasifikacija']
X = df.drop(columns=['Naziv_proteina', 'Klasifikacija'], axis=1)
```

```python
X.shape
```

```python
y.shape
```

```python
feature_names = X.columns
df[feature_names]
```

```python
y.unique()
```

```python
y.replace("uređen", 1, inplace=True)
y.replace("neuređen", 0, inplace = True)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
from sklearn.model_selection import GridSearchCV

params = { 'criterion' : ['gini', 'entropy'],
           'max_depth' : [2,4,6,8]
         }

model = GridSearchCV(DecisionTreeClassifier(), params, scoring = 'accuracy', cv=5, verbose=2)
```

```python
model.fit(X_train, y_train)
```

```python
model.best_score_
```

```python
model.best_params_
```

```python
best_model = model.best_estimator_
```

```python
best_model
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
y_pred_train = model.predict(X_train) 
confusion_matrix(y_train, y_pred_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(7, 7))
plot_tree(best_model, class_names=['uređen', 'neuređen'], feature_names=feature_names, filled=True)
plt.show()
```
