```python
#KOD NE RADI!!!
import pandas as pd
import matplotlib.pyplot as plt

# Učitavanje tabele sa podacima
df = pd.read_csv("ISUNSTRUCT_KNN_predikcije_YP_138523.1.csv") 

# Specifične dužine koje želimo da analiziramo 
specifikovane_duzine = [5, 6, 7]

# Funkcija za izračunavanje poklapanja za određene dužine
def izracunaj_poklapanje(df, duzine_list):
    poklapanje_za_duzine = []

    # Prolazimo kroz svaku traženu duzinu
    for duzina in duzine_list:
        # Filtriramo podatke za trenutnu duzinu
        filtrirani_podaci = df[df['Duzina'] == duzina]
        
        # Inicijalizacija broja poklapanja i ukupnog broja
        poklapanje = 0
        ukupan_broj = 0

        # Prolazimo kroz svaki redak za trenutnu duzinu
        for i in range(len(filtrirani_podaci)):
            knn_pred = filtrirani_podaci.iloc[i]['knn_pred']
            isunstruct_pred = filtrirani_podaci.iloc[i]['isunstruct_pred']
            
            # Poređenje predikcija
            if knn_pred == isunstruct_pred:
                poklapanje += 1
            ukupan_broj += 1

        # Izračunavanje procenta poklapanja
        if ukupan_broj > 0:
            procenat_poklapanja = (poklapanje / ukupan_broj) * 100
        else:
            procenat_poklapanja = 0  # Ako nema podataka za ovu duzinu

        # Dodavanje rezultata za trenutnu duzinu
        poklapanje_za_duzine.append((duzina, procenat_poklapanja))
    
    return poklapanje_za_duzine

# Izračunaj poklapanje za specifične dužine (npr. 5, 6, 7)
poklapanje = izracunaj_poklapanje(df, specifikovane_duzine)

# Prikazivanje rezultata
print("Procenti poklapanja za specifične dužine:", poklapanje)

# Sačuvaj rezultate u CSV fajl
result_df = pd.DataFrame(poklapanje, columns=["Duzina", "Procenat_poklapanja"])
result_df.to_csv("procenat_poklapanja_specifikuje_duzine.csv", index=False)

# Kreiranje grafika sa granicom
plt.figure(figsize=(10, 5))

# Plotujemo procenat poklapanja
plt.plot(result_df['Duzina'], result_df['Procenat_poklapanja'], marker='o', color='b', label='Poklapanje predikcija')

# Dodajemo horizontalnu liniju za granicu 50%
plt.axhline(y=50, color='r', linestyle='--', label='Granica 50%')

# Obojavamo područje ispod 50% kao 'uređeno' (u plavoj boji)
plt.fill_between(result_df['Duzina'], 0, 50, color='lightblue', alpha=0.5, label='Uređeno')

# Obojavamo područje iznad 50% kao 'neuređeno' (u crvenoj boji)
plt.fill_between(result_df['Duzina'], 50, result_df['Procenat_poklapanja'], color='salmon', alpha=0.5, label='Neuređeno')

# Postavljanje naslova i oznaka
plt.title('Procenat poklapanja između KNN i ISUNSTRUCT predikcija')
plt.xlabel('Duzina sekvence')
plt.ylabel('Procenat poklapanja (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Prikazivanje grafikona
plt.show()
```
