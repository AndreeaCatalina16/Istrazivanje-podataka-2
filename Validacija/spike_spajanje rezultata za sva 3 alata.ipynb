```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Uitaj CSV fajlove za VSL2, ISUNSTRUCT i IUPred3
df_vsl2 = pd.read_csv("/home/pc/Desktop/Istrazivanje-podataka-2/Validacija/VSL2_predikcije/Spike glycoprotein/procenti_za_KNN.csv")  # Pretpostavljam da je ovo CSV za VSL2
df_isunstruct = pd.read_csv("/home/pc/Desktop/Istrazivanje-podataka-2/Validacija/ISUNSTRUCT_predikcije/procenti_za_KNN.csv")  # CSV za ISUNSTRUCT
df_iupred3 = pd.read_csv("/home/pc/Desktop/Istrazivanje-podataka-2/Validacija/IUPRED3_predikcije/Spike glycoprotein/procenti_za_KNN.csv")  # CSV za IUPred3

# Dodaj kolonu sa imenom modela za grupisanje
df_vsl2['Model'] = 'VSL2'
df_isunstruct['Model'] = 'ISUNSTRUCT'
df_iupred3['Model'] = 'IUPred3'

# Spoji sve tri tabele u jednu
df_combined = pd.concat([df_vsl2, df_isunstruct, df_iupred3], ignore_index=True)

# Podesi stil grafika
sns.set_style("whitegrid")

# Kreiraj stubiasti grafikon
plt.figure(figsize=(12, 7))
ax = sns.barplot(x="Varijante", y="Procenat", hue="Model", data=df_combined, palette="Set2")

# Dodaj oznake na vrh stubaca
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}%", 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight="bold")

# Dodaj naslove i oznake
plt.xlabel("Kombinacija oznaka", fontsize=12)
plt.ylabel("Procenat (%)", fontsize=12)
plt.title("Predikcije spike glycoproteina dobijenih KNN modelom razliitim alatima (VSL2, ISUNSTRUCT, IUPRED3)", fontsize=14, fontweight="bold")
plt.xticks(rotation=45)

# Prikai grafikon
plt.show()
```
