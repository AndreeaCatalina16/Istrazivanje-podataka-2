```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_vsl2 = pd.read_csv("/home/pc/Desktop/Istrazivanje-podataka-2/Validacija/VSL2_predikcije/Spike glycoprotein/procenti_za_KNN.csv") 
df_isunstruct = pd.read_csv("/home/pc/Desktop/Istrazivanje-podataka-2/Validacija/ISUNSTRUCT_predikcije/procenti_za_KNN.csv") 
df_iupred3 = pd.read_csv("/home/pc/Desktop/Istrazivanje-podataka-2/Validacija/IUPRED3_predikcije/Spike glycoprotein/procenti_za_KNN.csv")  

df_vsl2['Alat'] = 'VSL2'
df_isunstruct['Alat'] = 'ISUNSTRUCT'
df_iupred3['Alat'] = 'IUPred3'

# Spoji sve tri tabele u jednu
df_combined = pd.concat([df_vsl2, df_isunstruct, df_iupred3], ignore_index=True)

sns.set_style("whitegrid")

# Kreiraj stubičasti grafikon
plt.figure(figsize=(12, 7))
ax = sns.barplot(x="Varijante", y="Procenat", hue="Alat", data=df_combined, palette="Set2")

for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}%", 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight="bold")

plt.xlabel("Kombinacija oznaka", fontsize=12)
plt.ylabel("Procenat (%)", fontsize=12)
plt.title("Predikcije spike glycoproteina dobijenih KNN modelom različitim alatima (VSL2, ISUNSTRUCT, IUPRED3)", fontsize=14, fontweight="bold")
plt.xticks(rotation=45)

plt.text(0.5, -0.32, "* VSL2 i IUPRED3 razmatraju sekvence dužine od 30 do 100", ha='center', va='center', fontsize=12, transform=plt.gca().transAxes, color='red')

# Prikaži grafikon
plt.show()
```
